```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from PIL import Image
import cv2

```

### Preprocessing - Skip to next section to load processed array that is to be used for the training dataset

```python
#Create pandas dataframe of training csv
train = pd.read_csv('archive/sign_mnist_train/sign_mnist_train.csv')
```

```python
#Create separate numpy arrays for pixel values and labels
pixels = train.iloc[:, 1:785].to_numpy()
labels = train.iloc[:, [0]].to_numpy()
```

```python
np.save("labels",labels)
```

```python
#View the first image
img = np.reshape(pixels[0][0:784],(28,28))
plt.imshow(img, cmap='gray')
```

```python
#Initialize empty array and fill with pixel data in 28x28 array
arr = np.empty([27455,28,28], int)

for i in range(len(pixels)):
    arr[i,:] = np.reshape(pixels[i][0:784],(28,28))
```

```python
#View first image
plt.imshow(arr[0], cmap='gray')
# same as img above
```

```python
#Resize the images to 224 x 224 
resized_images = np.array([Image.fromarray(x.astype(np.uint8)).resize([224,224]) for x in arr])
```

```python
#View the first image
plt.imshow(resized_images[0], cmap='gray')
```

```python
#Convert back to array
back_to_array = np.empty([27455,224,224], int)
for i in range(len(resized_images)):
    back_to_array[i] = np.asarray(resized_images[i])
    
```

```python
# View the first image
plt.imshow(back_to_array[0], cmap='gray')
```

```python
#Evaluate the array. The type needs to be np.uint8 to use opencv.
print(type(back_to_array[0][0][0]))
print(back_to_array[0][0][0])
```

```python
#Create another array of np.uint8 values
another_arr = np.empty([27455,224,224], np.uint8)
for i in range(len(back_to_array)):
    for j in range(len(back_to_array[i])):
        for k in range(len(back_to_array[i][j])):
            another_arr[i][j][k]=back_to_array[i][j][k].astype(np.int8)


```

```python
#Now correct type
print(type(another_arr[0][0][0]))
print(another_arr[0].shape)
print(another_arr[0][0][0])
```

```python
# View the first image
plt.imshow(another_arr[0], cmap='gray')
```

```python
#Save array in case something goes wrong
np.save("asl_stuff", another_arr)
```

### Part 2: neural network

```python
train_dataset = tf.data.Dataset.from_tensor_slices((truly_final_arr, labels))
# test_dataset = tf.data.Dataset.from_tensor_slices(())
```

```python
BATCH_SIZE = 64
SHUFFLE_BUFFER_SIZE = 100

train_dataset = train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
# test_dataset = test_dataset.batch(BATCH_SIZE)
```

```python
IMG_SIZE = (224, 224)
IMG_SHAPE = IMG_SIZE + (3,)

# Load base model
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top=False,
                                               weights='imagenet')

base_model.trainable = False
```

```python
# Create custom layers
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
# units here equals number of classes
prediction_layer = tf.keras.layers.Dense(units=26, activation='softmax')
identity_layer = tf.keras.layers.Layer()

inputs = tf.keras.Input(shape=(224, 224, 3))
x = preprocess_input(inputs)
x = base_model(x, training=False)
x = identity_layer(x)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
```

```python
##Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Train the model
initial_epochs = 10

model.fit(train_dataset,epochs=initial_epochs)
```
